```python
import pandas as pd
import numpy as np
from pathlib import Path 
#from matplotlib.pyplot as plt
#from scipy.stats import linregress
from ucimlrepo import fetch_ucirepo
```

```python
#Uploading CSV file from Cleveland

data_file1 = Path("heart+disease/processed.cleveland.data")
data_file1 = pd.read_csv(data_file1, header=None)
data_file1.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
data_file1["country"] = "cleveland"
data_file1
```

```python
#Uploading CSV file from Switzerland

data_file2 = Path("heart+disease/processed.switzerland.data")
data_file2 = pd.read_csv(data_file2, header=None)
data_file2.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
data_file2["country"] = "switzerland"
data_file2
```

```python
#Uploading CSV file from Va

data_file3 =  Path("heart+disease/processed.va.data")
data_file3 = pd.read_csv(data_file3, header=None)
data_file3.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
data_file3["country"] = "hungarian"
data_file3
```

```python
#Uploading CSV file from Hungarian

data_file4 =  Path("heart+disease/reprocessed.hungarian.data")
data_file4 = pd.read_csv(data_file4, header=None)
data_file4 = data_file4[0].str.split(' ', expand=True)
if data_file4.shape[1] == 14:
    data_file4.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
                           'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']
    data_file4["country"] = "hungarian"
else:
    print("The number of columns after splitting does not match the expected count.")
    
data_file4
```

```python
#combaining all 4 datataset together and creating a dataframe

dataframes = [data_file1, data_file2, data_file3, data_file4]
combined_df = pd.concat(dataframes, axis=0, ignore_index = True)
combined_df
```

```python
#checking what type of datas we have here

combined_df.info()
```

```python
#checking how many '?' are in the data

total_question_marks = (combined_df == '?').sum().sum()
print(total_question_marks)
```

```python
#checking how many rows with '?' are in the data

rows_with_question_mark = combined_df.isin(['?']).any(axis=1).sum()
print(rows_with_question_mark)
```

```python
#replacing '?'with 'NaN'

combined_df.replace('?', np.nan, inplace=True)
combined_df
```

```python
#checking how many duplicates are in the data

duplicate_count = combined_df.duplicated().sum()
print(f'The number of duplicate rows is: {duplicate_count}')
```

```python
#droping duplicates

df_cleaned = combined_df.drop_duplicates(keep='first')
df_cleaned
```

```python
#creating a list with colums that are numbers and changing their type to numeric, because
# we have checked that datas in these columns were objects

our_good_columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']


df_cleaned[our_good_columns] = df_cleaned[our_good_columns].apply(pd.to_numeric, errors='coerce')

df_cleaned

data_types = df_cleaned.dtypes
data_types
```

```python
# Our final data - we drop rows with values that were >0, because that's mean these values 
# are incorrect. 

final_clean_data = df_cleaned[(df_cleaned[our_good_columns] >= 0).all(axis=1)]
final_clean_data
```
